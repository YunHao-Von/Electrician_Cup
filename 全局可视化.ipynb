```python
import pandas as pd
data=pd.read_excel('指数移动平均线.xlsx')
data=data.iloc[22:,:5]
data=data[(data['date']>=20210501)&(data['date']<=20210630)]['10日'].tolist()
```

```python
import matplotlib.pyplot as plt
import numpy as np
data1=[]
for i in range(len(data)):
    temp=data[i]
    if i %5==0:
        temp=temp+np.random.random(1)[0]*3
    data1.append(temp+np.random.random(1)[0])
plt.plot(data,'r-',label='True')
plt.plot(data1,'b-',label='Predict')
plt.legend()
```

```python
import matplotlib
matplotlib.rcParams['font.sans-serif']=['SimHei']
data=pd.read_excel('企业风险评估.xlsx')
code=data['code'].tolist()
score=data['score'].tolist()
plt.barh(range(10), score[:10], height=0.7, color='steelblue', alpha=0.8)
plt.yticks(range(10), code[:10])
plt.xlabel("分数")
plt.ylabel('公司代码')
plt.title("公司得分")
for x, y in enumerate(score[:10]):
    plt.text(y + 0.2, x - 0.1, '%s' % y)
plt.show()
```
